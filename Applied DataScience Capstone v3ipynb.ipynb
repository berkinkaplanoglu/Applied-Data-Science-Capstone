```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data = pd.read_html(url)
data = data[0]
print(data)
```

```python
coorCsv = 'http://cocl.us/Geospatial_data'
Coords = pd.read_csv(coorCsv)
Coords
```

```python
data2 = data2.sort_values(by = 'Postal Code')
data2 = data2.reset_index()[['Postal Code','Borough','Neighbourhood']]
data2
```

```python
Coords2 = Coords[['Latitude','Longitude']]
Coords2
```

```python
newData = pd.concat([data2,Coords2],axis = 1)
newData
```

Let me tell you how i preprocessed the data. First i started with transforming data from wiki page. I extracted the tables at html page. Web page returned list of tables and our data was at first index of list. The table has 180 rows and 3 columns. I dropped columns that have 'Not assigned' value at Borough column. This method dropped 77 rows from dataset. I didn't apply any other method to preproccess data and our new dataset has 103 rows and 3 columns. Because of i don't want to deal with unreliable package. I used the other csv file that provided by IBM.  I reordered our neighbor dataset by Postal Code column and merged two files into one. Thats all i did before i continue.

```python
latitude = 43.72534265655438
longitude = -79.37259628709921
```

I used borough column to draw points at map.

```python
map = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(newData['Latitude'], newData['Longitude'], newData['Borough']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)  
    
map
```
