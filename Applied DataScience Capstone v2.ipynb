```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
data = pd.read_html(url)
data = data[0]
print(data)
```

```python
data2 = data[data['Borough'] != 'Not assigned']
```

```python
coorCsv = 'http://cocl.us/Geospatial_data'
Coords = pd.read_csv(coorCsv)
Coords
```

```python
data2 = data2.sort_values(by = 'Postal Code')
data2 = data2.reset_index()[['Postal Code','Borough','Neighbourhood']]
data2
```

```python
Coords2 = Coords[['Latitude','Longitude']]
Coords2
```

```python
newData = pd.concat([data2,Coords2],axis = 1)
newData
```
