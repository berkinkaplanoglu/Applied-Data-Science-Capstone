```python
CLIENT_ID = '***' # your Foursquare ID
CLIENT_SECRET = '***' # your Foursquare Secret
ACCESS_TOKEN = '***' # your FourSquare Access Token
VERSION = '20180604'
LIMIT = 60
```

```python
beslongitude = 29.00471116797716
beslatitude = 41.04351919765141


kadlongitude = 28.98449075402573
kadlatitude = 41.03722701768164 

search_query = 'lokma'
radius = 1000
```

Longitude and latitude of beşiktaş and taksim square.

```python
urlbes = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&oauth_token={}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, beslatitude, beslongitude,ACCESS_TOKEN, VERSION, search_query, radius, LIMIT)
urlkad = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&oauth_token={}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, kadlatitude, kadlongitude,ACCESS_TOKEN, VERSION, search_query, radius, LIMIT)
```

foursquare request link

```python
resultsbes = requests.get(urlbes).json()
resultskad = requests.get(urlkad).json()
```

```python
venuesbes = resultsbes['response']['venues']
dataframebes = json_normalize(venuesbes)

venueskad = resultskad['response']['venues']
dataframekad = json_normalize(venueskad)
```

transforming data into dataframe

```python
filtered_columnsbes = ['name', 'categories'] + [col for col in dataframebes.columns if col.startswith('location.')] + ['id']
dataframe_filteredbes = dataframebes.loc[:, filtered_columnsbes]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filteredbes['categories'] = dataframe_filteredbes.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filteredbes.columns = [column.split('.')[-1] for column in dataframe_filteredbes.columns]

dataframe_filteredbes
```

```python
filtered_columnskad = ['name', 'categories'] + [col for col in dataframekad.columns if col.startswith('location.')] + ['id']
dataframe_filteredkad = dataframekad.loc[:, filtered_columnskad]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
dataframe_filteredkad['categories'] = dataframe_filteredkad.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filteredkad.columns = [column.split('.')[-1] for column in dataframe_filteredkad.columns]

dataframe_filteredkad
```

```python
venues_mapbes = folium.Map(location=[41.01004605490622, 29.02376367217885], zoom_start=13)
for lat, lng, label in zip(dataframe_filteredbes.lat, dataframe_filteredbes.lng, dataframe_filteredbes.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(venues_mapbes)
    
for lat, lng, label in zip(dataframe_filteredkad.lat, dataframe_filteredkad.lng, dataframe_filteredkad.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='red',
        popup=label,
        fill = True,
        fill_color='red',
        fill_opacity=0.6
    ).add_to(venues_mapbes)

# display map
venues_mapbes
```
